# Exploratory Data Analysis

### Gender

```python
ax=sns.countplot(x='Gender', data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sales_gen
```

```python
sns.barplot(x='Gender', y='Amount', data=sales_gen)
```

### Age

```python
ax=sns.countplot(data=df, x='Age Group', hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age= df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sales_age
```

```python
v=sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

### State

```python
# Number of Orders for Top 10 States 
sales_state= df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders').head(10)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data=sales_state, x='State', y='Orders')
```

```python
# Total sales for Top 5 States

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount').head()

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data = sales_state, x='State',y='Amount')
```

### Marital status

```python
#CountPlot
ax = sns.countplot(data= df, x='Marital_Status')

sns.set(rc={'figure.figsize':(5,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total sales for Marital satus and gender
sales_status= df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(data = sales_status, x='Marital_Status', y='Amount', hue='Gender')
```

### Occupation

```python
#CountPlot
sns.set(rc={'figure.figsize':(20,5)})
ax= sns.countplot(data=df, x='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Occupation wise sales
sales_occ = df.groupby(['Occupation'], as_index=False)["Amount"].sum().sort_values(by='Amount')

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_occ, x='Occupation', y='Amount')
```

### Product Category

```python
#CountPlot
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df, x='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Sales for Top 10 Products
sales_prd = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_prd, x='Product_Category', y='Amount')
```

```python
# Total orders for Top 10 products
orders_pd = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10) 

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=orders_pd, x='Product_ID', y='Orders')
```
